````python
import tensorflow as tf
from PIL import Image
import matplotlib.pyplot as plt
import os

#이미지 불러오기
sample_img_path = os.getenv('HOME')+'/weekly-deeplearning-study/lecture_week1/data_augmentation/images/mycat.jpg'
sample_img_path
````

````python
# 이미지 리사이징 500x400
image = Image.open(sample_img_path).resize((500, 400)) # 이미지에 따라 숫자를 바꾸어 보세요.
image_tensor = tf.keras.preprocessing.image.img_to_array(image)

image
````

### Flip 해보기
---
Flip을 적용

````python
flip_lr_tensor = tf.image.flip_left_right(image_tensor) # 좌우반전
flip_ud_tensor = tf.image.flip_up_down(image_tensor) # 상하반전
flip_lr_image = tf.keras.preprocessing.image.array_to_img(flip_lr_tensor) # tensor to img
flip_ud_image = tf.keras.preprocessing.image.array_to_img(flip_ud_tensor) # tensor to img

plt.figure(figsize=(15, 15))

# 원본 사진
plt.subplot(1,3,1)
plt.title('Original image')
plt.imshow(image)

# 좌우반전 적용
plt.subplot(1,3,2)
plt.title('flip_left_right')
plt.imshow(flip_lr_image)

# 상하반전 적용
plt.subplot(1,3,3)
plt.title('flip_up_down')
plt.imshow(flip_ud_image)
````

랜덤으로 적용하기

````python
plt.figure(figsize=(12, 16))

row = 4
for i in range(row):
    flip_lr_tensor = tf.image.random_flip_left_right(image_tensor) # 랜덤 좌우 반전
    flip_ud_tensor = tf.image.random_flip_up_down(image_tensor) # 랜덤 상하 반전
    flip_lr_image = tf.keras.preprocessing.image.array_to_img(flip_lr_tensor)
    flip_ud_image = tf.keras.preprocessing.image.array_to_img(flip_ud_tensor)
    
    plt.subplot(4,3,i*3+1)
    plt.title('Original image')
    plt.imshow(image)

    plt.subplot(4,3,i*3+2)
    plt.title('flip_left_right')
    plt.imshow(flip_lr_image)

    plt.subplot(4,3,i*3+3)
    plt.title('flip_up_down')
    plt.imshow(flip_ud_image)
````

## 텐서플로우를 사용한 Image Augmentation (2) Center Crop

* `central_fraction`은 확대를 조절하는 매개변수. 1.0인 경우는 원본 이미지와 같다. 

````python
plt.figure(figsize=(12, 15))

central_fractions = [1.0, 0.75, 0.5, 0.25, 0.1]
col = len(central_fractions)
for i, frac in enumerate(central_fractions):
    cropped_tensor = tf.image.central_crop(image_tensor, frac) # Center Crop 적용
    cropped_img = tf.keras.preprocessing.image.array_to_img(cropped_tensor)
    
    plt.subplot(1,col+1,i+1)
    plt.title(f'Center crop: {frac}')
    plt.imshow(cropped_img)
````

* `centeral_crop`을 랜덤하게 적용하는 함수는 텐서플로우에서 제공되지 않음. 
* 파이썬의 `random` 모듈이나 텐서플로우의 랜덤 모듈 사용가능.

````python
# tf.random.uniform을 사용해서 central_fraction 매개변수를 정하고 crop하는 함수

def random_central_crop(image_tensor, range=(0, 1)):
    central_fraction = tf.random.uniform([1], minval=range[0], maxval=range[1], dtype=tf.float32)
    cropped_tensor = tf.image.central_crop(image_tensor, central_fraction)
    return cropped_tensor
````

````python
plt.figure(figsize=(12, 15))

col = 5
for i, frac in enumerate(central_fractions):
    cropped_tensor =random_central_crop(image_tensor)
    cropped_img = tf.keras.preprocessing.image.array_to_img(cropped_tensor)
    
    plt.subplot(1,col+1,i+1)
    plt.imshow(cropped_img)
````

__텐서플로우의 `tf.random.uniform()`과 `tf.random.normal()`의 차이.__

* 랜덤값을 uniform distribution(균일 분포)으로 뽑는 것과 normal distribution(정규 분포)으로 뽑는 차이. normal distribution의 경우 mean, std를 통해서 분포를 조절.

### tf.image.random_crop()
---
* [tf.image.random_crop](https://www.tensorflow.org/api_docs/python/tf/image/random_crop)

<center><b>들어간 인자들</b></center>

|value(input_tensor)|size(crop_size)|[seed](https://www.tensorflow.org/api_docs/python/tf/random/set_seed)|name|
|----|----|----|---|
|image_tensor|[180,180,3]|미정|미정|

````python
# apply random_crop on cat image
plt.figure(figsize=(12, 15))

random_crop_tensor = tf.image.random_crop(image_tensor,[180,180,3]) #input tensor, 이미지 사이즈
random_crop_image = tf.keras.preprocessing.image.array_to_img(random_crop_tensor)

plt.subplot(1,3,1)
plt.imshow(random_crop_image)
````

### tf.image.random_brightness()
---
* [tf.image.random_brightness](https://www.tensorflow.org/api_docs/python/tf/image/random_brightness)

<center><b>들어간 인자들</b></center>

|image(input_tensor)|max_delta(float 음수 불가)|seed(A Python integer)|
|----|----|----|
|image_tensor|[180,180,3]|미정|

````python
# apply random_brightness on cat image

random_bright_tensor = tf.image.random_brightness(image_tensor, max_delta=128)
random_bright_tensor = tf.clip_by_value(random_bright_tensor, 0, 255)
random_bright_image = tf.keras.preprocessing.image.array_to_img(random_bright_tensor)
plt.imshow(random_bright_image)
````

* `tf.image.random_brightness()` 다음에는 `tf.clip_by_value()`를 적용해주어야 함.
* `tf.image.random_brightness()`만 적용할 경우 이미지 텐서 값의 범위가 0~255를 초과하게 될 수도 있기 때문.

적용한 샘플 5장

````python
# display 5 random brightness images

plt.figure(figsize=(12, 15))

col = 5
for i in range(5):
    random_bright_tensor = tf.image.random_brightness(image_tensor, max_delta=128)
    random_bright_tensor = tf.clip_by_value(random_bright_tensor, 0, 255)
    random_bright_image = tf.keras.preprocessing.image.array_to_img(random_bright_tensor)

    plt.subplot(1,col+1,i+1)
    plt.imshow(random_bright_image)
````
